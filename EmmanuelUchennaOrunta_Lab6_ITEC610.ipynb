TASK 1

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("ITEC610_Lab_Dataset.csv")

# Set 'country' as the index
df_country = df.set_index('country')


# Using .loc to select countries with a population over 50 million
df_population_over_50m = df_country.loc[df_country['population'] > 50000000]

# Using .iloc to select first 5 countries listed alphabetically
df_sorted_ascending = df_country.sort_index()
df_first_5_countries = df_sorted_ascending.iloc[:5]

df_first_5_countries
```

TASK 2

```python
# Create a hierarchical MultiIndex using 'continent' and 'country'
df_multi_indexing = df.set_index(['continent', 'country'])

# Retrieve all countries in Africa starting with 'B' or 'C'
countries_africa_bc = df_multi_indexing.loc['Africa']
africa_bc_filtered = countries_africa_bc[countries_africa_bc.index.get_level_values('country').str.startswith(('B', 'C'))]

# Sort dataset by ascending continent and descending country
df_multi_sorted = df_multi_indexing.sort_index(level=['continent', 'country'], ascending=[True, False])

# Display both DataFrames clearly
print("African Countries Starting with B or C:")
display(africa_bc_filtered)

print("\nSorted MultiIndex DataFrame (Top Rows):")
display(df_multi_sorted.head())


```

TASK 3

```python
# Task 3: Data Integrity and Index Management

# Reset index to default numeric
df_reset = df_multi_indexing.reset_index()

# Assign 'life_expectancy' as the new index, sorted from highest to lowest
df_life_exp_indexed = df_reset.set_index('life_expectancy').sort_index(ascending=False)

# Using .loc to find countries with life expectancy above 75
life_exp_above_75 = df_life_exp_indexed.loc[df_life_exp_indexed.index > 75]

life_exp_above_75.head()
```
